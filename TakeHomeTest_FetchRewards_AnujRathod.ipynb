## Cleaning the "_id" from receipts dataset

```python
#There are "_id" to be cleaned in multiple dataset, below is the function to clean the "_id" column

def clean_id(column, dataset):   
    variable = []
    for i in range(0, len(dataset)):
        for key, value in column[i].items():
            variable.append(value)
            
            
    return variable
```

```python
receipts['_id'] = clean_id(receipts['_id'],receipts) #Clean id from receipts table 
```

## Cleaning the "dates" from receipts dataset

```python
#There are multiple "date" column to be cleaned in multiple dataset, below is the function to clean the "date" column

def clean_date(column, dataset): #fuction to clean date in the dataset
    
    var = []
    
    for i in range(0,len(dataset)):
        if(type(column[i])) == dict:
            for key, value in column[i].items():
                var.append(value)
        else:
            var.append(None)
        
    return var
```

```python
createDate = clean_date(receipts['createDate'],receipts)
receipts['createDate'] = pd.to_datetime(createDate, unit = 'ms')

dateScanned = clean_date(receipts['dateScanned'],receipts)
receipts['dateScanned'] = pd.to_datetime(dateScanned, unit = 'ms')

finishedDate = clean_date(receipts['finishedDate'],receipts)
receipts['finishedDate'] = pd.to_datetime(finishedDate, unit = 'ms')

modifyDate = clean_date(receipts['modifyDate'],receipts)
receipts['modifyDate'] = pd.to_datetime(modifyDate, unit = 'ms')

pointsAwardedDate = clean_date(receipts['pointsAwardedDate'],receipts)
receipts['pointsAwardedDate'] = pd.to_datetime(pointsAwardedDate, unit = 'ms')

purchaseDate = clean_date(receipts['purchaseDate'],receipts)
receipts['purchaseDate'] = pd.to_datetime(purchaseDate, unit = 'ms')
```

```python
#Making all the column names in lower case in order to write sql query
receipts.columns= receipts.columns.str.lower()
```

## Taking a temporary df to create a subset of receipts dataset, so we can flatten the "rewardsreceiptitemlist" column. this will open up the dictionary.

```python
df = receipts[['_id','rewardsreceiptitemlist']]
```

```python
out = []
for n, row in df.iterrows():
    if (type(row['rewardsreceiptitemlist'])) == list:
        for item in row['rewardsreceiptitemlist']:
            row['flatten'] = item
            out += [row.copy()]
    else:
        out+= [row.copy()]


flattened_df = pd.DataFrame(out)
```

```python
flattened_df

#From the below output I can see that there are list of dictionaries in rewardsreceiptitemlist column 
#(Example: Interpreting that "_id" for index 1 has two dictionaries)
```

```python
receipts = pd.merge(receipts,flattened_df, on = '_id') #Merging the flattened_df to my original receipts
```

```python
#Since we have flattened_df we no longer need "rewardsreceiptitemlist", so dropping it from the dataset
receipts.drop('rewardsreceiptitemlist_x', inplace = True, axis = 1)
receipts.drop('rewardsreceiptitemlist_y', inplace = True, axis = 1)
```

```python
#Renaming flatten as "rewardsreceiptitemlist" 
receipts.rename(columns = {'flatten':'rewardsreceiptitemlist'}, inplace = True)
```

```python
#The code below will provide unique list of keys in all the dictionary (This is our column for the new table to be created)  

all_keys = set()

for i in range(len(receipts)):
    if (type(receipts['rewardsreceiptitemlist'][i])) == dict:
        all_keys = all_keys.union(set(receipts['rewardsreceiptitemlist'][i].keys()))
```

```python
all_keys #These are all the unique keys from the 'rewardsreceiptitemlist' column
```

```python
rewards_receipt = pd.DataFrame(columns = ['_id']+list(all_keys))  #Structure of new table to be created
rewards_receipt
```

```python
# Giving all the value to the respected key to the new structured table. Null will be assigned to the key if they dont have values

columns = ['_id']+list(all_keys)
dictionary = dict()


for each in columns:
    dictionary[each] = []
    
for i in range(len(receipts)):
    dictionary['_id'].append(receipts['_id'].values[0])
    tempDic = receipts['rewardsreceiptitemlist'][i]
    
    for each in all_keys:
        if (type(tempDic)) == dict:
            dictionary[each].append(tempDic.get(each,np.nan)) 
        else:
            dictionary[each].append(np.nan)
```

```python
rewards_receipt = pd.DataFrame(dictionary)
rewards_receipt
```

```python
#Converting all the column names in lower case in order to write sql query
rewards_receipt.columns= rewards_receipt.columns.str.lower()
```

```python
#"rewardsreceiptitemlist" column is no longer needed in the original receipts dataset
#Since a new table rewards_receipt was created from this rewardsreceiptitemlist column

receipts.drop('rewardsreceiptitemlist', inplace = True, axis = 1)
```
